```python
import numpy as np
import pandas as pd
import random
```

```python
train=pd.read_csv("../input/sign-language-mnist/sign_mnist_train.csv")
train.info()
```

```python
y_inp=train.iloc[:1000,0]
x_inp=train.iloc[:1000,1:]
x_train=np.array(x_inp)
y_train=pd.get_dummies(y_inp.values)
y_train=np.array(y_train)
x_train=x_train/255
```

```python
def sigmoid(x):
        return 1.0/(1.0+np.exp(-x))
```

```python
w1=np.random.rand(784,784)
b1=list()
for i in range(784):
    x=np.random.randint(-300,-200)
    b1.append(x)
b1=np.array(b1)
b1=b1.reshape(784,1)
```

```python
w2=np.random.rand(784,24)
b2=list()
for i in range(24):
    x=np.random.randint(-250,-100)
    b2.append(x)
b2=np.array(b2)
b2=b2.reshape(24,1)
```

```python
for i in range(10):
    print(i)
    error=[]
    for i in range(1000):
        x=x_train[i].reshape(784,1)
        l1=np.dot(w1,x)
        l1=l1.reshape(784,1)
        l1+=b1
        s1=sigmoid(l1) 
        l2=np.dot(w2.T,s1)
        l2=l2.reshape(24,1)
        l2+=b2
        z=sigmoid(l2)
        truth=y_train[i].reshape(24,1)
        error.append((truth-z)**2)
        a=np.dot((1-z),z.T)
        b=np.dot(l1,(truth-z).T)
        #dW2=((1+z)*z.T)@(l1(truth-z)*T)
        dW2=np.dot(b,a)
        w2-=dW2
        t1=np.dot(np.dot((truth-z),z.T),(1-z))
        t2=np.dot(np.dot(w2,t1),x.T)
        dW1=np.dot(np.dot(1-s1,s1.T),t2)
        w1-=dW1
error=np.array(error)
error.shape
```

```python
error.mean()
```
